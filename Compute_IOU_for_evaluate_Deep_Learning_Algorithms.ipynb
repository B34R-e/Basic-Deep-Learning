```python
def computeIOU(boxA, boxB):
  # determine the (x, y) cordinates of the intersection rectangle
  xA = max(boxA[0], boxB[0])
  yA = max(boxA[1], boxB[1])
  xB = min(boxA[2], boxB[2])
  yB = min(boxA[3], boxB[3])

  # compute the intersection area
  intersection_Area = max(0, xB - xA + 1) * max(0, yB - yA + 1)
  # compute the area of both boxA and boxB (prediction and ground-truth box)
  boxA_Area = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
  boxB_Area = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
  # compute the intersection over union (IOU) using Intersection Area divided by Union Area
  # Union Area equals sum of (BoxA Area, BoxB Area) minus the Intersection Area
  IOU = intersection_Area / ((boxA_Area + boxB_Area) - intersection_Area)
  return IOU
```

```python
# Example:
boxA = (0, 0, 5, 5)
boxB = (2.5, 2.5, 7.5, 7.5)
IOU = computeIOU(boxA, boxB)
print(IOU)
```
